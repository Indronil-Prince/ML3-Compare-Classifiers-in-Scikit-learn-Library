```python
# Required Libraries
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Perceptron, LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import warnings
warnings.filterwarnings('ignore') 
```

**Dataset 1: Digits Dataset from Scikit-Learn Datasets**

```python
# Load digits dataset
digits = datasets.load_digits()
```

```python
# Splitting data
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
```

```python
# Data preprocessing
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Perceptron
print("Perceptron Model:")
perceptron = Perceptron(eta0=0.01, random_state=42)
start_time = time.time()
perceptron.fit(X_train, y_train)
end_time = time.time()
time_perceptron = end_time - start_time
print(f"Training Time: {time_perceptron:.4f} seconds")
y_pred_train = perceptron.predict(X_train)
y_pred_test = perceptron.predict(X_test)
train_accuracy_perceptron = accuracy_score(y_train, y_pred_train)
test_accuracy_perceptron = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {train_accuracy_perceptron:.2f}")
print(f"Test Accuracy: {test_accuracy_perceptron:.2f}")
```

```python
# Logistic Regression
print("Logistic Regression Model:")
log_reg = LogisticRegression(C=0.1, random_state=42)
start_time = time.time()
log_reg.fit(X_train, y_train)
end_time = time.time()
time_log_reg = end_time - start_time
print(f"Training Time: {time_log_reg:.4f} seconds")
y_pred_train = log_reg.predict(X_train)
y_pred_test = log_reg.predict(X_test)
train_accuracy_log_reg = accuracy_score(y_train, y_pred_train)
test_accuracy_log_reg = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {train_accuracy_log_reg:.2f}")
print(f"Test Accuracy: {test_accuracy_log_reg:.2f}")
```

```python
# Linear SVM
print("Linear SVM Model:")
linear_svm = SVC(kernel='linear', C=1.0, random_state=42)
start_time = time.time()
linear_svm.fit(X_train, y_train)
end_time = time.time()
time_linear_svm = end_time - start_time
print(f"Training Time: {time_linear_svm:.4f} seconds")
y_pred_train = linear_svm.predict(X_train)
y_pred_test = linear_svm.predict(X_test)
train_accuracy_linear_svm = accuracy_score(y_train, y_pred_train)
test_accuracy_linear_svm = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {train_accuracy_linear_svm:.2f}")
print(f"Test Accuracy: {test_accuracy_linear_svm:.2f}")
```

```python
# Non-linear SVM with RBF Kernel
print("Non-linear SVM with RBF Kernel Model:")
non_linear_svm = SVC(kernel='rbf', C=1.0, gamma=0.01, random_state=42)
start_time = time.time()
non_linear_svm.fit(X_train, y_train)
end_time = time.time()
time_non_linear_svm = end_time - start_time
print(f"Training Time: {time_non_linear_svm:.4f} seconds")
y_pred_train = non_linear_svm.predict(X_train)
y_pred_test = non_linear_svm.predict(X_test)
train_accuracy_non_linear_svm = accuracy_score(y_train, y_pred_train)
test_accuracy_non_linear_svm = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {train_accuracy_non_linear_svm:.2f}")
print(f"Test Accuracy: {test_accuracy_non_linear_svm:.2f}")
```

```python
# Decision Tree
print("Decision Tree Model:")
dt = DecisionTreeClassifier(max_depth=10, random_state=42)
start_time = time.time()
dt.fit(X_train, y_train)
end_time = time.time()
time_dt = end_time - start_time
print(f"Training Time: {time_dt:.4f} seconds")
y_pred_train = dt.predict(X_train)
y_pred_test = dt.predict(X_test)
train_accuracy_dt = accuracy_score(y_train, y_pred_train)
test_accuracy_dt = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {train_accuracy_dt:.2f}")
print(f"Test Accuracy: {test_accuracy_dt:.2f}")
```

```python
# KNN
print("KNN Model:")
knn = KNeighborsClassifier(n_neighbors=5)
start_time = time.time()
knn.fit(X_train, y_train)
end_time = time.time()
time_knn = end_time - start_time
print(f"Training Time: {time_knn:.4f} seconds")
y_pred_train = knn.predict(X_train)
y_pred_test = knn.predict(X_test)
train_accuracy_knn = accuracy_score(y_train, y_pred_train)
test_accuracy_knn = accuracy_score(y_test, y_pred_test)
print(f"Train Accuracy: {train_accuracy_knn:.2f}")
print(f"Test Accuracy: {test_accuracy_knn:.2f}")
```

```python
# Accuracy values for each classifier
train_accuracies = [train_accuracy_perceptron, train_accuracy_log_reg, train_accuracy_linear_svm, train_accuracy_non_linear_svm, train_accuracy_dt, train_accuracy_knn]
test_accuracies = [test_accuracy_perceptron, test_accuracy_log_reg, test_accuracy_linear_svm, test_accuracy_non_linear_svm, test_accuracy_dt, test_accuracy_knn]
classifiers = ['Perceptron', 'Logistic Regression', 'Linear SVM', 'Non-linear SVM (RBF)', 'Decision Tree', 'KNN']

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

# Plot the accuracy values
ax[0].bar(classifiers, train_accuracies, color='skyblue')
ax[0].set_xlabel('Classifiers')
ax[0].set_ylabel('Time (seconds)')
ax[0].set_xticklabels(classifiers, rotation=60, ha='right')
ax[0].set_title('Time to Train Each Classifier')

ax[1].bar(classifiers, test_accuracies, color='orange')
ax[1].set_xlabel('Classifiers')
ax[1].set_ylabel('Time (seconds)')
ax[1].set_xticklabels(classifiers, rotation=60, ha='right')
ax[1].set_title('Time to Test Each Classifier')
plt.show()
```

```python
# Time values for each classifier
time_values = [time_perceptron, time_log_reg, time_linear_svm, time_non_linear_svm, time_dt, time_knn]
classifiers = ['Perceptron', 'Logistic Regression', 'Linear SVM', 'Non-linear SVM (RBF)', 'Decision Tree', 'KNN']

# Plot the time values
plt.figure(figsize=(10,4))
plt.barh(classifiers, time_values, color='skyblue')
plt.xlabel('Time (seconds)')
plt.ylabel('Classifiers')
plt.title('Time to Train Each Classifier')
plt.show()
```

```python
# Parameter Grids
param_grid_perceptron = {'eta0': [0.01, 0.1, 0.3, 0.5, 0.7, 1.0]}
param_grid_log_reg = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
param_grid_svm_linear = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
param_grid_svm_rbf = {'gamma': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
param_grid_dt = {'max_depth': np.arange(1, 21)}
param_grid_knn = {'n_neighbors': np.arange(1, 21)}
```

```python
# Tuning Perceptron
perceptron = Perceptron(random_state=42)
grid_search_perceptron = GridSearchCV(estimator=perceptron, param_grid=param_grid_perceptron, cv=5, return_train_score=True)
grid_search_perceptron.fit(X_train, y_train)
best_eta0_perceptron = grid_search_perceptron.best_params_['eta0']
print(f"Best eta for Perceptron: {best_eta0_perceptron}")

# Tuning Logistic Regression
log_reg = LogisticRegression(random_state=42)
grid_search_log_reg = GridSearchCV(estimator=log_reg, param_grid=param_grid_log_reg, cv=5, return_train_score=True)
grid_search_log_reg.fit(X_train, y_train)
best_C_log_reg = grid_search_log_reg.best_params_['C']
print(f"Best C for Logistic Regression: {best_C_log_reg}")

# Tuning SVM (Linear)
svm_linear = SVC(kernel='linear', random_state=42)
grid_search_svm_linear = GridSearchCV(estimator=svm_linear, param_grid=param_grid_svm_linear, cv=5, return_train_score=True)
grid_search_svm_linear.fit(X_train, y_train)
best_C_svm_linear = grid_search_svm_linear.best_params_['C']
print(f"Best C for Linear SVM: {best_C_svm_linear}")

# Tuning SVM (RBF)
svm_rbf = SVC(kernel='rbf', random_state=42)
grid_search_svm_rbf = GridSearchCV(estimator=svm_rbf, param_grid=param_grid_svm_rbf, cv=5, return_train_score=True)
grid_search_svm_rbf.fit(X_train, y_train)
best_gamma_svm_rbf = grid_search_svm_rbf.best_params_['gamma']
print(f"Best gamma for SVM (RBF): {best_gamma_svm_rbf}")

# Tuning Decision Tree
dt = DecisionTreeClassifier(random_state=42)
grid_search_dt = GridSearchCV(estimator=dt, param_grid=param_grid_dt, cv=5, return_train_score=True)
grid_search_dt.fit(X_train, y_train)
best_max_depth_dt = grid_search_dt.best_params_['max_depth']
print(f"Best max_depth for Decision Tree: {best_max_depth_dt}")

# Tuning KNN
knn = KNeighborsClassifier()
grid_search_knn = GridSearchCV(estimator=knn, param_grid=param_grid_knn, cv=5, return_train_score=True)
grid_search_knn.fit(X_train, y_train)
best_n_neighbors_knn = grid_search_knn.best_params_['n_neighbors']
print(f"Best n_neighbors for KNN: {best_n_neighbors_knn}")
```

```python
# Plotting Accuracy vs. Hyperparameters
def plot_hyperparameter_tuning(param_grid, best_params, param_name, accuracy_train, accuracy_test, title):
    plt.figure(figsize=(10, 5))
    plt.plot(param_grid[param_name], accuracy_train, label='Train Accuracy')
    plt.plot(param_grid[param_name], accuracy_test, label='Test Accuracy')
    plt.scatter(best_params[param_name], max(accuracy_test), marker='x', color='red', label='Best Hyperparameter')
    plt.xlabel(param_name)
    plt.ylabel('Accuracy')
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Plotting Perceptron Hyperparameter Tuning
plot_hyperparameter_tuning(param_grid_perceptron, grid_search_perceptron.best_params_, 'eta0', 
                            grid_search_perceptron.cv_results_['mean_train_score'], 
                            grid_search_perceptron.cv_results_['mean_test_score'], 
                            'Perceptron: Accuracy vs. eta')

# Plotting Logistic Regression Hyperparameter Tuning
plot_hyperparameter_tuning(param_grid_log_reg, grid_search_log_reg.best_params_, 'C', 
                            grid_search_log_reg.cv_results_['mean_train_score'], 
                            grid_search_log_reg.cv_results_['mean_test_score'], 
                            'Logistic Regression: Accuracy vs. C')

# Plotting SVM (Linear) Hyperparameter Tuning
plot_hyperparameter_tuning(param_grid_svm_linear, grid_search_svm_linear.best_params_, 'C', 
                            grid_search_svm_linear.cv_results_['mean_train_score'], 
                            grid_search_svm_linear.cv_results_['mean_test_score'], 
                            'SVM (Linear): Accuracy vs. C')

# Plotting SVM (RBF) Hyperparameter Tuning
plot_hyperparameter_tuning(param_grid_svm_rbf, grid_search_svm_rbf.best_params_, 'gamma', 
                            grid_search_svm_rbf.cv_results_['mean_train_score'], 
                            grid_search_svm_rbf.cv_results_['mean_test_score'], 
                            'SVM (RBF): Accuracy vs. gamma')

# Plotting Decision Tree Hyperparameter Tuning
plot_hyperparameter_tuning(param_grid_dt, grid_search_dt.best_params_, 'max_depth', 
                            grid_search_dt.cv_results_['mean_train_score'], 
                            grid_search_dt.cv_results_['mean_test_score'], 
                            'Decision Tree: Accuracy vs. max_depth')

# Plotting KNN Hyperparameter Tuning
plot_hyperparameter_tuning(param_grid_knn, grid_search_knn.best_params_, 'n_neighbors', 
                            grid_search_knn.cv_results_['mean_train_score'], 
                            grid_search_knn.cv_results_['mean_test_score'], 
                            'KNN: Accuracy vs. n_neighbors')
```
